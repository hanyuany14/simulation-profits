```python
import numpy as np
import pandas as pd

from main import Main

np.random.seed(0)

# 輸入參數
salvage_value = 0
cost = 400
price = 1000

time_period = 10
simulation_times = 1000  # 代表每一個 Qk 包含多少筆
items_num_train = 20  # 有幾個 items in training
items_num_test = 5  # 有幾個 items in testing

# input data
demand_df_train = (
    np.random.normal(loc=15, scale=3, size=(items_num_train, time_period))
    .round()
    .astype(int)
)
demand_df_train = pd.DataFrame(demand_df_train)
print(f"demand_df_train.shape: {demand_df_train.shape}")
# print(demand_df_train)


demand_df_test = (
    np.random.normal(loc=15, scale=3, size=(items_num_test, time_period))
    .round()
    .astype(int)
)
demand_df_test = pd.DataFrame(demand_df_test)
print(f"demand_df_test.shape: {demand_df_test.shape}")
# print(demand_df_test)


Qks_test = np.random.normal(
    loc=20, scale=5, size=(time_period - 2, items_num_test, simulation_times)
)
print(f"Qks_test.shape: {Qks_test.shape}")


# 執行主程式
app = Main()

(
    train_profit_df,
    training_stimulation_result_df,
    training_results,
) = app.main(
    cost=cost,
    price=price,
    salvage_value=salvage_value,
    time_period=time_period,
    items_num_train=items_num_train,
    items_num_test=items_num_test,
    simulation_times=simulation_times,
    Qks_test=Qks_test,
    demand_df_train=demand_df_train,
    demand_df_test=demand_df_test,
)
```
